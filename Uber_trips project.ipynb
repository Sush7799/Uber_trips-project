```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/DataScience75/Batch-59-Datasets/main/Uber_raw_2019.csv'
```

```python
df = pd.read_csv(url, index_col = 0)
```

```python
df.head()
```

```python
df.columns
```

```python
df.index
```

```python
df = df.reset_index()
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date/Time'] = df['Date/Time'].map(pd.to_datetime)
```

```python
df.info()
```

```python
df['Day'] = pd.to_datetime(df['Date/Time']).dt.day
```

```python
df.head()
```

```python
df['Hour'] = pd.to_datetime(df['Date/Time']).dt.hour
```

```python
df.head()
```

```python
df['Weekday'] = pd.to_datetime(df['Date/Time']).dt.weekday
```

```python
df.head()
```

```python
df['Month'] = pd.to_datetime(df['Date/Time']).dt.month
```

```python
df.head()
```

```python
df['Year'] = pd.to_datetime(df['Date/Time']).dt.year
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
sns.set(rc = {'figure.figsize':(12,10)})
sns.distplot(df['Day'])
```

```python
### by this we can show that on 4th,5th and 6th trips demand are high.
```

```python
sns.set(rc = {'figure.figsize':(12,10)})
sns.distplot(df['Weekday'])
```

```python
sns.heatmap(df1, annot = False)
```

```python
sns.scatterplot(x = 'Lat', y = 'Lon', hue = 'Weekday', data = df)
```

```python
df['Date'] = df['Date/Time'].dt.date
df
```

```python
df2 = df.groupby('Date').size().reset_index().rename(columns = {0: 'No. of trips'})
df2
```

```python
### date wise count of trips
```

```python
df2['MovingAvg_trips'] = df2['No. of trips'].rolling(2).mean()
```

```python
df2
```

```python
df2[['No. of trips','MovingAvg_trips']].plot(label="Trips", figsize = (16,8))
```

```python
df2['Cumulative_trips'] = df2['No. of trips'].expanding().mean()
df2.head()
```

```python
df2[['No. of trips','Cumulative_trips','MovingAvg_trips']].plot(label="Trips", figsize = (16,8))
```

```python
df2['Exp_trips'] = df2['No. of trips'].ewm(span = 2).mean()
```

```python
df2.head()
```

```python
df2[['No. of trips','Cumulative_trips','MovingAvg_trips','Exp_trips']].plot(label="Trips", figsize = (16,8))
```
